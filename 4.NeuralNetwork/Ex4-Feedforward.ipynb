```
import numpy as np
```

```
def weight(shape): return np.ones(shape)
def bias(shape): return np.ones(shape)
def relu(x): return x * (x > 0)
```

```
input_shape = 2
#  nodes in the hidden layer 1 & 2
nodes = 3
output_shape = 1
```

```
#  the input layer
input = np.random.uniform(size=input_shape)
w1 = weight((input_shape, nodes))
b1 = bias(nodes)
z1 = np.matmul(input, w1) + b1
a1 = relu(z1)

# the hidden layers
w2 = weight((nodes, nodes))
b2 = bias(nodes)
z2 = np.matmul(a1, w2) + b2
a2 = relu(z2)

# the output layer
w3 = weight((nodes, output_shape))
b3 = bias(output_shape)
z3 = np.matmul(a2, w3) + b3
a3 = relu(z3)


```

```
print('input {}'.format(input))
print('input layer {}'.format(a1))
print('hidden layer  {}'.format(a2))
print('output {}'.format(a3))
```
