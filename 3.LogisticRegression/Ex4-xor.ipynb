```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
N = 4
D = 2

X = np.array([[0, 0],[0, 1],[1, 0],[1, 1],])
y = np.array([0, 1, 1, 0]).reshape(-1, 1)
ones = np.array([[1]*N]).T

plt.scatter(X[:,0], X[:,1], c=y)

plt.show()
```

```
# Hàm sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```
# Thêm cột 1 vào dữ liệu x
x = np.hstack((np.ones((N, 1)), X))

w = np.array([0.,0.1,0.1,0.1]).reshape(-1,1)
print(w.shape)
```

```
x = np.hstack((x,([0],[1],[1],[0])))
x
```

```
# Số lần lặp bước 2
numOfIteration = 10000
cost = np.zeros((numOfIteration,1))
learning_rate = 0.1

w = np.array([0.,0.1,0.1,0.1]).reshape(-1,1)

for i in range(1, numOfIteration):
    
	 # Tính giá trị dự đoán
    y_predict = sigmoid(np.dot(x, w))
    cost[i] = -np.sum(np.multiply(y, np.log(y_predict)) + np.multiply(1-y, np.log(1-y_predict)))
    # Gradient descent
    # np.dot: nhan ma tran, * / multiply : element-wise
    # print(w.shape, x.shape, y_predict.shape, y.shape, (y_predict-y).shape)
    w = w - learning_rate * np.dot(x.T, y_predict-y)	 
   
```
