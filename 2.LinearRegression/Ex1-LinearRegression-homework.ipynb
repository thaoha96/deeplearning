# 1.Implement thuật toán gradient descent tìm giá trị nhỏ nhất hàm f(x)=x^2, thử các giá trị learning_rate khác nhau (nhỏ, vừa, lớn, rất lớn), vẽ đồ thị loss function theo epoch tương ứng.



```
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

def function(x):
  return (x**2)

def deriv(x):
  return 2*x

learning_rates = [0.0000001,0.05,0.001,0.1]
def GD(l_r):
  numOfIteration = 100
  x0 = -2
  x=[x0]
  cost = np.zeros((numOfIteration,1))
  for i in range(0,numOfIteration):
    x_new = x[-1]-l_r*deriv(x[-1])
    x.append(x_new)
    cost[i] = (function(x[i]))
  print( np.amin(cost))
  return cost


for num,l_r in enumerate(learning_rates):
  loss = GD(l_r)
  plt.plot(loss, label=str(l_r))
plt.xlabel("epoch#")
plt.ylabel("loss")
plt.legend()
plt.show()
```
