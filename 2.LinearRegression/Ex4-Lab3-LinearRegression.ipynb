```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = pd.read_csv('data_square.csv').values
N = data.shape[0]
x = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)
plt.scatter(x, y)
plt.xlabel('mét vuông')
plt.ylabel('giá')
```

```
x = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)
x = np.stack([x, np.square(x)], axis=-1)
x = x.reshape(x.shape[0], x.shape[-1])
print(x.shape)
x
```

```
x = np.hstack((np.ones((N, 1)), x))
w = np.array([0,1,1]).reshape(-1,1)
x
```

```
numOfIteration = 10000
cost = np.zeros((numOfIteration,1))
learning_rate = 0.0000000001
for i in range(0, numOfIteration):
    r = np.dot(x, w) - y
    cost[i] = 0.5*np.sum(r*r)*(1/N)
    # correct the shape dimension
    w = w - learning_rate * np.dot(x.T, r)*(1/N)
    print('step {}, cost: {}'.format(i, cost[i]))
print( np.amin(cost))
```

```
plt.plot(cost)
```
