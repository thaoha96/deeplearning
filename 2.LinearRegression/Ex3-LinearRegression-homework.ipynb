```
data = pd.read_csv('data_linear.csv').values
N = data.shape[0]
x = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)

plt.scatter(x, y)
plt.xlabel('mét vuông')
plt.ylabel('giá')
```

```
data[:, 0].reshape(-1, 1)
```

```
np.ones((N, 1))
```

```
x = np.hstack((np.ones((N, 1)), x))

w = np.array([0.,1.]).reshape(-1,1)

x
```

```

#learning_rate = 0.000001
def GD(x,y,w,learning_rate):
  numOfIteration = 100
  cost = np.zeros((numOfIteration,1))
  for i in range(0, numOfIteration):
      r = np.dot(x, w) - y
      cost[i] = 0.5*np.sum(r*r)*(1/N)
      # correct the shape dimension
      w = w - learning_rate * (1/N)* np.dot(x.T, r)
  #   print('step {}, cost: {}'.format(i, cost[i]))
  return cost

learning_rates = [0.000001,0.00001]
for num,l_r in enumerate(learning_rates):
  loss = GD(x,y,w,l_r)
  plt.plot(loss)
  plt.xlabel("epoch#")
  plt.ylabel("loss")
```

```
learning_rates2 = [0.001,1]
for num,l_r in enumerate(learning_rates2):
  loss = GD(x,y,w,l_r)
  plt.plot(loss, labels=str(l_r))
plt.xlabel("epoch#")
plt.ylabel("loss")
plt.legend()
plt.show()
```

```
numOfIteration = 100
cost = np.zeros((numOfIteration,1))
learning_rate = 0.000001
for i in range(0, numOfIteration):
    r = np.dot(x, w) - y
    cost[i] = 0.5*np.sum(r*r)
    # correct the shape dimension
    w = w - learning_rate * np.dot(x.T, r)
    print('step {}, cost: {}'.format(i, cost[i]))
```
